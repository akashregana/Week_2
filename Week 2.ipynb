```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as mpl
```

```python
df = sb.load_dataset("titanic")
```

```python
#Histograms - distribution of numerical features
sb.histplot(data=df, x='age', bins=30, kde=True)
mpl.title("Age Distribution")
mpl.show()
```

```python
#Boxplots - to spot outliers
sb.boxplot(x=df['fare'])
mpl.title("Fare Boxplot")
mpl.show()
```

```python
#Correlation Matrix
correlation = df.corr(numeric_only=True)
sb.heatmap(correlation, annot=True, cmap="coolwarm")
mpl.title("Correlation Matrix")
mpl.show()
```

```python
#High fares
df[df['fare'] > 300]
#Very old passengers
df[df['age'] > 70]
```

```python
sb.countplot(data=df, x='sex', hue='survived')
mpl.title("Survival Count by Gender")
mpl.show()
```

```python
#Data Cleaning 
df.dropna(subset=['age', 'fare', 'sex', 'survived'], inplace=True)
```

```python
#Visual EDA
#age vs survival
sb.histplot(data=df, x='age', hue='survived', bins=30, kde=True)
mpl.title("Age vs Survival")
mpl.show()
```

```python
#fare vs survival
sb.boxplot(x='survived', y='fare', data=df)
mpl.title("Fare vs Survival")
mpl.show()
```

```python
#gender vs survival
sb.countplot(data=df, x='sex', hue='survived')
mpl.title("Gender vs Survival")
mpl.show()
```

```python
#statical Summary
print(df.describe())
print(df['sex'].value_counts())
print(df['survived'].value_counts(normalize=True))
```
